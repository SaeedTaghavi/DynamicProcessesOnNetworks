# Complex Systems

The material we will be studying in this course falls under the general concept of **Complex Systems**.  What does it mean to say we are studying a complex system?  It's a somewhat nebulous concept.  In a sense, it's something that when we've had enough experience we know it when we see it.  Of course, that's not useful for someone who is being introduced to the concepts.  So here is my best attempt at a definition: 

> *A **Complex System** is a system made up of autonomous objects (that is, there is no external coordination) whose combined actions produce some **emergent** system-scale outcome*

This definition still lacks some information because I've used the word "*emergent*", without clearly defining it.

> *An outcome is **emergent** if it is not an obvious combination of the actions of the individual objects.* 

For example, if we know how one object behaves, if the behavior of $N$ objects is simply $N$ times as large, they are not a complex system.

Notice that the definition of a complex system requires both the objects and the outcome.  A system may not be a complex system if we are interested in one quantity, while it is a complex system if we are interested in another.  

Unfortunately, these definitions are not as absolute as mathematics definitions usually are.  The mathematical models we develop later will be more precise.

# Network Science

Notice that many if not all of the complex systems examples we have considered depend on pairwise interactions.  The structure of interactions is fundamentally important to the outcomes.  Often the structure of interactions is more important than the properties of the objects.  This motivates the concept of a **network**.  

> *A **network** is a collection of interacting objects that are represented as **nodes**, and interactions between the objects are treated as **edges**. *

Networks may change over time, but for our purposes we will assume they change more slowly than whatever process we are studying.  So we treat the network as static.  This means we are neglecting many important possibilities, such as networks whose edges evolve in response to whatever is happening on the network.

The definition of a network is closely related to the definition of a graph.  Indeed a network is basically a graph.  The terminology reflects different research communities studying the same objects, but with different emphasis.  In *network science*, we are usually interested in properties of processes that happen on the network, while in *graph theory* we are usually interested in the properties of the graph itself.

### The Karate Club:

In the early 1970s, [Wayne Zachary studied the social interactions between individuals in a karate club](https://en.wikipedia.org/wiki/Zachary%27s_karate_club).  The observed friendships form a small graph that is widely used as an example in network science.

We're going to look at the club and observe some simple properties using *Python* and *networkx*.  I am consciously showing you this before we have a separate introduction to Python and networkx.  This of course means that I don't expect you to be able to produce the code below on your own.  But I will try to keep it simple enough that it still makes sense.

Here are the friendship interactions he identified, with two different views:

```python
import networkx as nx            #networkx provides python commands that will be useful for networks
import matplotlib.pyplot as plt  #matplotlib.pyplot provides commands that are useful for plotting

G = nx.karate_club_graph()       #read in the karate club graph which networkx has stored.

nx.draw_circular(G, with_labels=True)  #use the networkx command for drawing G with nodes in a circle.
plt.show()                             #display the plot (it will appear below)
plt.clf()                              #clear the figure (otherwise the following plot would also have this plotted)

pos = nx.spring_layout(G)       #find positions for the nodes treating the nodes as weights and edges as springs
                                #the algorithm starts with nodes randomly positioned and then moves them around trying
                                #to find an equilibrium position.
nx.draw(G, pos, with_labels=True)   #draw the graph with the positions just found
plt.show()
```

During his study, individuals 0 (the instructor) and 33 (the club administrator/president) had a dispute which split the club into two new clubs.  Based on the structure of this network, can we predict who was in each of the two new clubs?  Let's try.

```python
instructornodes = [0]
presidentnodes = [33]

#We can use set.append(element) to specify each node's predicted alignment.  
#Or we can use set.extend(list).  So for example

instructornodes.append(11)
presidentnodes.extend([14,15,20])

#do more here

plt.clf()

#now we can draw them, using some networkx command you probably don't know yet
nx.draw(G, pos, nodelist = instructornodes, node_color = 'blue')
nx.draw(G, pos, nodelist = presidentnodes, node_color = 'orange')
nx.draw_networkx_labels(G, pos)
#until you've added all the nodes, it'll be leaving a bunch blank
plt.show()

```

Here is the outcome that actually occurred:

```python
instructornodes = [node for node in G.nodes() if G.nodes[node]['club'] == 'Mr. Hi']
presidentnodes = [node for node in G.nodes() if G.nodes[node]['club'] == 'Officer']

plt.clf()
nx.draw(G, pos, nodelist = instructornodes, node_color = 'blue')
nx.draw(G, pos, nodelist = presidentnodes, node_color = 'orange')
nx.draw_networkx_labels(G, pos)

plt.show()
```

Clearly this is a reasonable way to expect the friendship network to split.  Notice that in everything above there, I didn't say anything about the fact that the dispute was over how much the lessons should cost.  Based only on the network structure can give a reasonable prediction about how the club would split without knowing why the club split or who (if anyone) was "right" or "wrong".  

There is a large amount of research that goes into identifying communities within networks, and for any new algorithm, the most basic test is whether it closely predicts the Karate Club split.

## Network Formalization

We've seen strong hints that the structure of a network is fundamental to the outcome of processes on that network.

In order to study this, we need some more definitions.  We're going to start by defining a network.  Since we'll be using *Networkx* in this course, we'll stick to its definitions.  There are four different types of networks we can use.  

> *A **network** can be *
>
> - *A **Graph**:  In a Graph, edges exist between pairs of nodes.  Any two nodes have at most one edge.  A node may have an edge with itself.  The edges have no direction.  These treat the partnerships as symmetric.*
>
> - *A **DiGraph**: A DiGraph is like a Graph, except that the edges have direction.  Two nodes can have zero edges, one edge in one direction, one edge in the other direction, or one edge in each direction.*
> 
> - *A **MultiGraph**: A MultiGraph is like a Graph, but with the restrictions that a pair of nodes may have more than one edge.*
> 
> - *A **MultiDiGraph**: A MultiDiGraph generalizes a DiGraph in the same way that a MultiGraph generalizes a Graph.  Repeated edges are allowed.*

The code below here demonstrates these network types.  I **do not** expect you to be familiar with the commands, or really understand the syntax yet.

Notice that these networks are **not** "simple graphs" because all of them allow self-edges and the multi(di)graphs allow repeated edges.  We could (and often do) assume that our networks do not have self-edges.

```python
import networkx as nx
G1 = nx.Graph()
G1.add_edge(1,2)
print('first print statement: ', G1.edges())  #there is one edge
G1.add_edge(1,2)   #the edge already exists.  Networkx ignores this command
print('second print statement: ', G1.edges())
G1.add_edge(2,1)   #in a Graph, a (2,1) edge is the same as (1,2).  Networkx ignores this command.
print('third print statement: ', G1.edges())
G1.add_edge(3,3)
print('fourth print statement: ', G1.edges())
```

```python
G2 = nx.DiGraph()
G2.add_edge(1,2)
print('first print statement: ', G2.edges())  #there is one edge
G2.add_edge(1,2)   #the edge already exists.  Networkx ignores this command
print('second print statement: ', G2.edges())
G2.add_edge(2,1)   #This is treated as a different edge
print('third print statement: ', G2.edges())
G2.add_edge(3,3)
print('fourth print statement: ', G2.edges())
```

```python
import networkx as nx
G3 = nx.MultiGraph()
G3.add_edge(1,2)
print('first print statement: ', G3.edges())  #there is one edge
G3.add_edge(1,2)   #the edge already exists.  Networkx adds another
print('second print statement: ', G3.edges())
G3.add_edge(2,1)   #A (2,1) edge is the same as (1,2).  Networkx adds another such edge.  Now there are 3
print('third print statement: ', G3.edges())
G3.add_edge(3,3)
print('fourth print statement: ', G3.edges())
```

```python
G4 = nx.MultiDiGraph()
G4.add_edge(1,2)
print('first print statement: ', G4.edges())  #there is one edge
G4.add_edge(1,2)   #the edge already exists.  Networkx adds another
print('second print statement: ', G4.edges())
G4.add_edge(2,1)   #This is treated as a different edge
print('third print statement: ', G4.edges())
G4.add_edge(3,3)
print('fourth print statement: ', G4.edges())
```

### Network Properties

We need to define a few important properties used to measure networks.

- **Degree**:  
The **degree** of a node is the number of edges it is involved in.  If the network is directed, we can define the **in-degree** as the number of edges pointing to a node and its **out-degree** as the number of edges pointing from the node.

- **Degree distribution**:  
The **degree distribution** of a network is the distribution of the degrees of the nodes.

- **Clustering**:  
The **clustering** of a graph is a measure of the frequency of triangles.  Unfortunately there are [**two similar but distinct definitions** of clustering](https://en.wikipedia.org/wiki/Clustering_coefficient) in a network that are widely used.  For this course, we will use the following:
  - The **clustering coefficient** of a graph is the probability that if we choose a random node $v$ and consider two of its neighbors $u$ and $w$, then $u$ and $w$ share an edge.  (this is the **network average clustering coefficient** as defined on the Wikipedia page).
  - The **transitivity** of a graph is the probability that if we choose a random triple $u$-$v$-$w$ then $u$ and $w$ share an edge.  (this is the **global clustering coefficient** as defined on the Wikipedia page).
 
 These are different because in the second case we choose the middle nodes based on how often they occur in a triple, so high degree nodes occur more often, while in the first case we choose the middle nodes with equal probability.

- **Node weights**:  
A node's **weight** is some attribute assigned to the node to reflect its relative importance for some process.  For example, in thinking about an infectious disease, we might weight a node more highly if it has not had a recent vaccine.

- **Edge weights**:  
Similarl to a node weight, an edge's weight is an attribute assigned to the edge to reflect its relative importance.  For an infectious disease, we might weight an edge more highly if the two individuals interact more frequently.

Some network properties can be demonstrated in this figure, borrowed from [Aaron Clauset's lecture notes](http://tuvalu.santafe.edu/~aaronc/courses/3352/):

![Sample Networks from Aaron Clauset's lecture notes](images/network_example.png)

In the left-hand network, node 3 has degree 4 and node 5 has degree 2.

The following code generates the left-hand network in networkx and reports some of the degrees:

```python
import networkx as nx  #note that if you've previously used this command in this notebook, 
                       #it's not needed now
import matplotlib.pyplot as plt
G = nx.Graph()
edgelist = [(1,2), (1,5), (2,3), (2,4), (3,4), (3,5), (3,6)]  #this is a list of edges
G.add_edges_from(edgelist)
print(G.degree(3))
print(G.degree(5))
plt.clf()
nx.draw_networkx(G)
plt.show()
```
